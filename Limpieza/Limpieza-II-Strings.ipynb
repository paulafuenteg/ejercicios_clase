## Creación de nuevas columnas basadas en columnas ya existentes

En algunas ocasiones nos podeomos encontrar columnas con *strings* que están bastante "sucias". Podemos tener columnas que contegan demasiada información y que queramos separar en distintas columnas, o crear intervalos en función de los valores de una columna o incluso podemos crear columnas que nos indiquen porcentajes o ratios. En este apartado veremos como podemos hacer todo esto de una forma muy sencilla usando Pandas 👯. 

### Separar una columna en dos

Exploremos la columna `latest_household_survey`.

````python
df["latest_household_survey"].unique()[:10]
````

Si nos fijamos esta columna hace referencia a cuando se hizo el último muestreo. Sin embargo, es una columna de tipo *string* y tiene más información a parte de los años. 

Lo que vamos a hacer es quedarnos únicamente con los números (con los años). Para ello nos vamos a usar el método propio de pandas `str.split()`.

Este método va a **dividir *strings* basándose en un separador/delimitador dado**.

Su sintaxis: 

```python
    columna.str.split(pat=None, n=- 1, expand=False, *, regex=None)
```

Donde: 

- `pat`: str o regex, opcional. Cadena o expresión regular para dividir. Si no se especifica, se divide en espacios en blanco. Si especificamos por `";"` dividiremos por `;`. 

- `n`: int, por defecto `-1` (todos). Limita el número de divisiones. `None`, `0` y `-1` serán interpretados como la devolución de todas las divisiones.

- `expand`: bool, por defecto `False`. Separa las strings divididas en columnas separadas.

    - Si es `True`, devuelve tantas columnas nuevas como divisiones.
    - Si es `False`, devuelve una única columna con una lista donde tendremos todas las divisiones. 

- `regex`:  

    - Si es `True`, asume que el patrón pasado es una expresión regular.
    - Si es `False`, trata el patrón como una cadena literal.


Cosas que debemos tener en cuenta para crear la función 

- Tenemos `nan` por lo que tendremos que hacer un `try`, `except` para evitar que nos de errores. 
- Los años vienen siempre después de una coma, por lo que tendremos que usar el método `split` de los *strings*.

````python
df[['survey_place','year_survey']] = df['latest_household_survey'].str.split(',', expand=True, n=1)
````

````python
df.sample(2)
````

Si nos fijamos se nos han creado dos columnas nuevas llamadas `survey_place` y `year_survey`, donde en la primera tendremos el sitio donde se hizo la encuesta y en la segunda los años👏🏽. 

### Creación de intervalos: `pd.cut()`

```python
    pd.cut(x, bins, right=True, labels=None, ordered=True)
```

Donde: 

- `x`: la columna que vamos a querer dividir. La matriz de entrada que se va a dividir. Debe ser unidimensional.

- `bins`: integer, el número de intervalos en el que queremos dividir nuestra columna. 

- `labels`: array o `False`, por defecto `None`.
Especifica las etiquetas de los intervalos devueltos. Deben tener la misma longitud que los intervalos resultantes. Si es `False`, devuelve sólo indicadores enteros de los bins. Si es `True`, genera un error si no coincide. Cuando `ordered=False`, se deben proporcionar las etiquetas.

- `ordered`: bool, por defecto `True`. Si las etiquetas están ordenadas o no. Si es `False`, la columna resultante no estará ordenado (se deben proporcionar las etiquetas).

````python
df["age_bins"] = pd.cut(df["mean_age"], 3, labels = ["grupoA", "grupoB", "grupoC"])
````

````python
df.sample(10)
````

````python
df["age_bins"].unique()
````

### Creación de porcentajes

````python
df.head(2)
````

En este apartado utilizaremos las columnas de `m2016` y `f2016`, las cuáles indican el número de defunciones y nacimientos respectivamente en cada pais. 

Para esto nos crearemos una función para calcular el porcentaje de nacidos en función de toda la población.

````python
def porcentaje(nacidos, total): 
    return (nacidos * 100) / total
````

````python
df["porcentaje_nacidos"] = df.apply(lambda datos: porcentaje(datos["f2016"], datos["total2016"]), axis = 1)
````

````python
df.head(2)
````

Lo podemos hacer de una forma más sencilla sin usar funciones:

````python
df["porcentaje_nacidos2"] = (df["f2016"] * 100) / df["total2016"]
````

````python
df.head(2)
````

### Creación de Ratios 

La diferencia entre ratio y porcentaje es mínima: 

- `ratio`:  es el cociente


- `porcentaje`: es el cociente multiplicado por cien (expresado como tanto por ciento).

En este caso calcularemos el ratio de nacidos, y usaremos una `lambda`. 


````python
df["ratio_nacidos"] = df.apply(lambda df: df["f2016"] / df["total2016"], axis = 1)
````

````python
df.head(2)
````

## Otras formas de modificar una columna o DataFrame al completo: 

En Pandas tenemos:

- `apply`: se puede aplicar tanto al DataFrame completo como a una sola columna, ya lo hemos visto.

- `pd.applymap`: se puede aplicar solo a todo el DataFrame.

- `map`: solo lo podremos aplicar a una columa en particular. 

Veamos los dos nuevos uno a uno: 

### `pd.applymap`

Aplica una función a un Dataframe por elementos, es decir, similar al `apply` que ya conocemos con la diferencia que este método se tiene que aplicar a **todo el dataframe**. 

Este método aplica una función que acepta y devuelve un escalar a cada elemento de un DataFrame.

``` python
    MiDataFrame.applymap(func, na_action=None, **kwargs)
```

Donde: 

- `func`: función de Python, devuelve un único valor a partir de un único valor.

- `na_action`: puede ser `None` o `'ignore'`, por defecto `None`.

    - Si es `'ignore'`, propaga los valores `NaN`, sin pasarlos a `func`. Es decir, ignora los nulos. 

- `**kwargs`: argumentos adicionales para pasar como argumentos de palabras clave a `func`.


````python
# definimos nuestra lambda: 

scaled_df=df.applymap(lambda a: a + 10 if type(a) == int or type(a) == float else a, na_action = "ignore")
````

````python
scaled_df.head(1)
````

````python
df.head(1)
````

Si nos fijamos por ejemplo en la columna `m2016` vemos que los valores hay incrementado en 10. Esto es porque en nuestra lambda hemos definido que aquellos que sean *integers* o *floats* les sumaremos 10. En caso contrario dejaremos los valores originales. 

````python
# si intentaramos aplicarlo a una columna: 

scaled_df=df["mean_age"].applymap(lambda a: a + 10 if type(a) == int or type(a) == float else a, na_action = "ignore")
````

Nos sale un error y nos dice que el método `applymap` no se puede aplicar a `Series`. Por lo tanto: 

- `applymap`: únicamente para todo el `DataFrame`.

- `apply`: lo podemos aplicar tanto a todo el `DataFrame` como a una columna en concreto (una `Serie`).

### `map`

Mapea los valores de la columna de acuerdo a un mapeo o función dada.

Se utiliza para sustituir cada valor de una columna por otro valor, que puede derivarse de una función, un `dict` o una `Serie`.

Su sintaxis: 

```python
    MiColumna.map(arg, na_action=None)
```

Donde: 

- `arg`: function para mapear.

- `na_action`: puede ser `None` o `'ignore'`, por defecto `None`.

    - Si es `'ignore'`, no tiene en cuenta los nulos.

En la columna `system_of_trade` tenemos dos valores únicos diferentes. 

```python
    df["system_of_trade"].unique()
```

output

```console
    array(['General trade system', 'Special trade system', nan], dtype=object)
```

Imaginemos que queremos mapear estos valores para ponerlos como *integers*. Podríamos crearnos una función y aplicarla a nuestra columna. Pero tenemos el método `map` que nos lo va a permitir hacer de una forma más sencilla. 

Aunque ahora no le veamos mucho sentido a esta tranformación, nos puede resultar super útil cuando lleguemos a la parte de *machine learning*, ya que muchos modelos no aceptan variables categóricas y las tendremos que cambiar a valores numéricos. Es en este momento donde tendremos que usar el `map`. 

Veamos un ejemplo con un diccionario, que es la forma más usual de usar el `map`. 

````python
# lo primero que tenemos que hacer es crear el diccionario para el mapeo
mapa = {"General trade system": 1, # donde los valores sean "General trade system" pondremos un 1
       "Special trade system": 2} # donde los valores sean "Special trade system" podremos un 0
````

````python
# ejecutamos la función map 
df["system_of_trade"] = df["system_of_trade"].map(mapa, na_action="ignore")
````

````python
# chequeamos los cambios. Ahora en la columna "system_of_trade" ya no tendremos los valores antiguos si no 1 y 2. 
df.head(1)
````

````python
df["system_of_trade"].unique()
````

````python
df.loc[28]
````

````python
# si intentaramos aplicar un map a toda un dataframe 
df_ = df.map(mapa, na_action="ignore")
````

Nos sale un error y nos dice que el método `map` no se puede aplicar a un `DataFrame`. Por lo tanto: 

- `applymap`: únicamente para todo el `DataFrame`.


- `apply`: lo podemos aplicar tanto a todo el `DataFrame` como a una columna en concreto (una `Serie`).


- `map`: únicamente para columnas.
