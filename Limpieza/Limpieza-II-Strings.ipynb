### Separar una columna en dos

Exploremos la columna `latest_household_survey`.

````python
df["latest_household_survey"].unique()[:10]
````

Si nos fijamos esta columna hace referencia a cuando se hizo el 煤ltimo muestreo. Sin embargo, es una columna de tipo *string* y tiene m谩s informaci贸n a parte de los a帽os. 

Lo que vamos a hacer es quedarnos 煤nicamente con los n煤meros (con los a帽os). Para ello nos vamos a usar el m茅todo propio de pandas `str.split()`.

Este m茅todo va a **dividir *strings* bas谩ndose en un separador/delimitador dado**.

Su sintaxis: 

```python
    columna.str.split(pat=None, n=- 1, expand=False, *, regex=None)
```

Donde: 

- `pat`: str o regex, opcional. Cadena o expresi贸n regular para dividir. Si no se especifica, se divide en espacios en blanco. Si especificamos por `";"` dividiremos por `;`. 

- `n`: int, por defecto `-1` (todos). Limita el n煤mero de divisiones. `None`, `0` y `-1` ser谩n interpretados como la devoluci贸n de todas las divisiones.

- `expand`: bool, por defecto `False`. Separa las strings divididas en columnas separadas.

    - Si es `True`, devuelve tantas columnas nuevas como divisiones.
    - Si es `False`, devuelve una 煤nica columna con una lista donde tendremos todas las divisiones. 

- `regex`:  

    - Si es `True`, asume que el patr贸n pasado es una expresi贸n regular.
    - Si es `False`, trata el patr贸n como una cadena literal.


Cosas que debemos tener en cuenta para crear la funci贸n 

- Tenemos `nan` por lo que tendremos que hacer un `try`, `except` para evitar que nos de errores. 
- Los a帽os vienen siempre despu茅s de una coma, por lo que tendremos que usar el m茅todo `split` de los *strings*.

````python
df[['survey_place','year_survey']] = df['latest_household_survey'].str.split(',', expand=True, n=1)
````

````python
df.sample(2)
````

Si nos fijamos se nos han creado dos columnas nuevas llamadas `survey_place` y `year_survey`, donde en la primera tendremos el sitio donde se hizo la encuesta y en la segunda los a帽os. 

### Creaci贸n de intervalos: `pd.cut()`

```python
    pd.cut(x, bins, right=True, labels=None, ordered=True)
```

Donde: 

- `x`: la columna que vamos a querer dividir. La matriz de entrada que se va a dividir. Debe ser unidimensional.

- `bins`: integer, el n煤mero de intervalos en el que queremos dividir nuestra columna. 

- `labels`: array o `False`, por defecto `None`.
Especifica las etiquetas de los intervalos devueltos. Deben tener la misma longitud que los intervalos resultantes. Si es `False`, devuelve s贸lo indicadores enteros de los bins. Si es `True`, genera un error si no coincide. Cuando `ordered=False`, se deben proporcionar las etiquetas.

- `ordered`: bool, por defecto `True`. Si las etiquetas est谩n ordenadas o no. Si es `False`, la columna resultante no estar谩 ordenado (se deben proporcionar las etiquetas).

````python
df["age_bins"] = pd.cut(df["mean_age"], 3, labels = ["grupoA", "grupoB", "grupoC"])
````

En este apartado utilizaremos las columnas de `m2016` y `f2016`, las cu谩les indican el n煤mero de defunciones y nacimientos respectivamente en cada pais. 

Para esto nos crearemos una funci贸n para calcular el porcentaje de nacidos en funci贸n de toda la poblaci贸n.

### Creaci贸n de Ratios 

La diferencia entre ratio y porcentaje es m铆nima: 

- `ratio`:  es el cociente


- `porcentaje`: es el cociente multiplicado por cien (expresado como tanto por ciento).

En este caso calcularemos el ratio de nacidos, y usaremos una `lambda`. 


````python
df["ratio_nacidos"] = df.apply(lambda df: df["f2016"] / df["total2016"], axis = 1)
````

## Otras formas de modificar una columna o DataFrame al completo: 

En Pandas tenemos:

- `apply`: se puede aplicar tanto al DataFrame completo como a una sola columna, ya lo hemos visto.

- `pd.applymap`: se puede aplicar solo a todo el DataFrame.

- `map`: solo lo podremos aplicar a una columa en particular. 

Veamos los dos nuevos uno a uno: 

### `pd.applymap`

Aplica una funci贸n a un Dataframe por elementos, es decir, similar al `apply` que ya conocemos con la diferencia que este m茅todo se tiene que aplicar a **todo el dataframe**. 

Este m茅todo aplica una funci贸n que acepta y devuelve un escalar a cada elemento de un DataFrame.

``` python
    MiDataFrame.applymap(func, na_action=None, **kwargs)
```

Donde: 

- `func`: funci贸n de Python, devuelve un 煤nico valor a partir de un 煤nico valor.

- `na_action`: puede ser `None` o `'ignore'`, por defecto `None`.

    - Si es `'ignore'`, propaga los valores `NaN`, sin pasarlos a `func`. Es decir, ignora los nulos. 

- `**kwargs`: argumentos adicionales para pasar como argumentos de palabras clave a `func`.


````python
# definimos nuestra lambda: 

scaled_df=df.applymap(lambda a: a + 10 if type(a) == int or type(a) == float else a, na_action = "ignore")
````

````python
scaled_df.head(1)
````

````python
df.head(1)
````

Si nos fijamos por ejemplo en la columna `m2016` vemos que los valores hay incrementado en 10. Esto es porque en nuestra lambda hemos definido que aquellos que sean *integers* o *floats* les sumaremos 10. En caso contrario dejaremos los valores originales. 

````python
# si intentaramos aplicarlo a una columna: 

scaled_df=df["mean_age"].applymap(lambda a: a + 10 if type(a) == int or type(a) == float else a, na_action = "ignore")
````

Nos sale un error y nos dice que el m茅todo `applymap` no se puede aplicar a `Series`. Por lo tanto: 

- `applymap`: 煤nicamente para todo el `DataFrame`.

- `apply`: lo podemos aplicar tanto a todo el `DataFrame` como a una columna en concreto (una `Serie`).

### `map`

Mapea los valores de la columna de acuerdo a un mapeo o funci贸n dada.

Se utiliza para sustituir cada valor de una columna por otro valor, que puede derivarse de una funci贸n, un `dict` o una `Serie`.

Su sintaxis: 

```python
    MiColumna.map(arg, na_action=None)
```

Donde: 

- `arg`: function para mapear.

- `na_action`: puede ser `None` o `'ignore'`, por defecto `None`.

    - Si es `'ignore'`, no tiene en cuenta los nulos.

En la columna `system_of_trade` tenemos dos valores 煤nicos diferentes. 

```python
    df["system_of_trade"].unique()
```

output

```console
    array(['General trade system', 'Special trade system', nan], dtype=object)
```

Imaginemos que queremos mapear estos valores para ponerlos como *integers*. Podr铆amos crearnos una funci贸n y aplicarla a nuestra columna. Pero tenemos el m茅todo `map` que nos lo va a permitir hacer de una forma m谩s sencilla. 

Aunque ahora no le veamos mucho sentido a esta tranformaci贸n, nos puede resultar super 煤til cuando lleguemos a la parte de *machine learning*, ya que muchos modelos no aceptan variables categ贸ricas y las tendremos que cambiar a valores num茅ricos. Es en este momento donde tendremos que usar el `map`. 

Veamos un ejemplo con un diccionario, que es la forma m谩s usual de usar el `map`. 

````python
# lo primero que tenemos que hacer es crear el diccionario para el mapeo
mapa = {"General trade system": 1, # donde los valores sean "General trade system" pondremos un 1
       "Special trade system": 2} # donde los valores sean "Special trade system" podremos un 0
````

````python
# ejecutamos la funci贸n map 
df["system_of_trade"] = df["system_of_trade"].map(mapa, na_action="ignore")
````

- `applymap`: 煤nicamente para todo el `DataFrame`.

- `map`: 煤nicamente para columnas.
