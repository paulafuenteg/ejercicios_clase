# Estructuras control de bucles

Las sentencias `break`,  `continue` y `pass` nos ayudan a afinar nuestros bucles y su ejecuci칩n.

- `break`: rompe el bucle por completo

- `continue`: salta el resto del bucle actual y pasa a la siguiente iteraci칩n

## `break`

Se utiliza en situaciones en las que queremos salir del bucle, incluso si la condici칩n no se ha convertido en False o hemos iterado sobre toda la secuencia. Adem치s, si se utiliza `break`, no se ejecutan los siguientes bloques del c칩digo.

Veamos algunos ejemplos, porque sobre esta teor칤a nos puede resultar un poco ambiguo. 

````python
# imaginemos que tenemos una lista de palabras

palabras = ['python', 'programacion', 'adalab', 'titanas']
````

Lo que vamos a querer es que si una de las palabras de la lista contiene una "a" el c칩digo se pare. 

````python
# si no incluyeramos un break en nuestro c칩digo 
for w in palabras:
    if "a" in w:
        print(f"La palabra {w} contiene la letra 'a'")
````

Como vemos el c칩digo de arriba nos esta devolviendo aquellas palabras que contienen la letra "a". Pero como hemos dicho, lo que nos interesa es que el c칩digo se pare cuando encuentre una palabra que contenga la letra "a". Para eso usaremos la sentencia `break`

````python
# vamos a para el bucle que cuando llegue a una palabra con 'e' se pare

for w in palabras:
    if "a" in w:
        print(f"La palabra {w} contiene la letra 'a'!!")
        break
````

쯈u칠 es lo que ha pasado? 

- Se ha iniciado el *for loop* iterando por nuestra lista de palabras

- Chequeamos la condicion de si la letra "a" est치 en cada una de las palabras de nuestra lista. 

- En caso de que la condici칩n anterior se cumpla, nos saldr치 un mensaje en pantalla 

- A continuaci칩n est치 el estamento `break` que no lo que har치 ser치 parar el c칩digo si se cumple la condici칩n. Al pararse el c칩digo no se sigue ejecutando y no veremos m치s palabras que contengan la "a". 

## `continue`

Parecido al `break` con que paramos la ejecuci칩n de un bucle tambi칠n existe la sentencia `continue`. En ese caso se deja de ejecutar lo dem치s del bucle `while` tambi칠n, pero con la diferencia que sigue con la siguiente iteraci칩n, en vez de parar por completo.

````python
i = 0
while i < 6:
  i += 1
  if i == 3:
    continue
  print(i)
````

## `pass`

La instrucci칩n `pass` permite manejar la condici칩n sin que el bucle se vea afectado de ninguna manera; todo el c칩digo continuar치 ley칠ndose a menos que se produzca la instrucci칩n `break` u otra instrucci칩n.

````python
lista_nombres = ['Mario','Alice', 'Bob', 'Jack', 'Mary', 'Manuel', 'Amanda', 'Ignacio', 'Fernando']
````

````python
for nombre in lista_nombres:
    if "o" in nombre:
        pass
    else:
        print("Hola", nombre)
````

````python
# veamos otro ejemplo con pass. En este caso vamos a iterar por por una lista de n칰meros. A cada n칰mero le vamos a sumar 10, pero si los numeros son igual a 5 o 7 no querremos que pase nada

# iniciamos el bucle
for num in range(10):

    # establecemos la condici칩n. En caso de que el n칰mero sea igual a 5 o 7 no queremos que pase nada
    if num == 5 or num == 7: 
        pass   

    # para todos los numeros que no cumplan la condici칩n, le sumaremos 10
    else:
        num += 10
        print(f'La suma del n칰mero es {num}')
````

Si nos fijamos, en los *prints* tenemos la suma de todos los elementos menos los del 5 y el 7, ya que hemos especificado en el c칩digo que si eran cualquiera de esos n칰meros no quer칤amos que pasara nada. 

**Repaso**
- Con `continue` se salta una iteraci칩n de un bucle `while`.
- Con `else:` se puede especificar qu칠 hacer al salir del `while`.

## `try: ... except`:

Su sintaxis es: 

```python

       # el c칩digo que queramos

       # c칩digo opcional
``` 


````python
# imaginemos que queremos hacer una lista con los resultados de la division de 2 entre una serie de numeros, entre ellos el 0, que pasar치?

# creamos la lista vac칤a para ir a침adiendo los resultados de la divisi칩n
divisiones2 = []

# empezamos nuestro for loop
for i in range(-3, 5):

    # a침adimos los resultado de la divisi칩n  en la lista vac칤a
    divisiones2.append(2 / i)
````

Vaya... nos devuelve un error, y es que no podemos dividir entre 0. Lo malo de esto es que no se ha podido ejecutar todo el c칩digo. Para estas situaciones son 칰tiles los `try: ... except:`. Lo que haremos ser치 decirle a nuestro c칩digo, intenta hacer esto, que no puedes... no pasa nada! Veamos un ejemplo: 

````python

# esta parte de c칩digo es igual que en la celda anterior
divisiones3 = []
for i in range(-3, 5):

    # con el try le decimos que intente hacer una parte del c칩digo
    try:
        divisiones3.append(2 / i)
    
    # en caso de que el c칩digo anterior de error y no se pueda hacer...
    except:
        # nos printea un error diciendo sobre que n칰mero no nos pudo ejecutar la l칤nea de error
        print(f'no se puede dividir entre {i}')

        # y apendeamos a la lista "no data"
        divisiones3.append("no data")
````

Si nos fijamos ahora nuestro programa no se par칩, nos sali칩 un mensaje de que no se pudo dividir entre 0 y en la lista de los resultados deber칤amos tener todos los resultados de las divisiones y un valor deber칤a ser "no datda". Veamoslo!
