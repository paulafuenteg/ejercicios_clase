Creemos un tablero de ajedrez. En este caso, nuestro array tendrá que tenemos 0 y 1 de forma alterna. Por lo que empezaremos con una matriz de ceros. 

```python
# creamos la matriz de 0
array = np.zeros([5,7], dtype = int)
array
```

Ahora necesitamos cambiar algunos ceros por unos. Tendremos que tener en cuenta los siguientes detalles: 

Tendremos que poner ceros en valores de columnas alternas y de filas

```python
# iteramos por nuestro array
for i in range(array.shape[0]):
    if i%2 == 0:  # chequeamos la condición para saber si la fila es par o impar. En caso de que sea par
       print("par")
       array[i, 1::2] = 1 # reemplazamos los valores de las filas pares y los elementos de las columas impares por 1,  que se cambiarán por blanco
    else:
        print("impar")
        array[i, 0::2] = 1 # columnas pares filas impares en blanco
        
    
```

```python
plt.imshow(array, cmap = "gray");
```
