# ¿Qué es Seaborn? <a class="anchor" id="section_1_1"></a>

Seaborn es una librería utilizada para realizar gráficos estadísiticos en Python. Está construida sobre las bases de Matplotlib y se integra de forma impecable con Pandas.

Lo cual nos abre una serie de puertas para facilitar la creación de gráficas complejas en muy pocas líneas de código. Asímismo tiene gráficas creadas exclusivamente para la visualización de variables categóricas.

Descubriremos que podemos realizar las gráficas realizadas en las lecciones anteriores en Matplotlib con Seaborn y explicaremos nuevos gráficos.

## Creación de gráficas en Seaborn <a class="anchor" id="section_1_3"></a>

Llegados a este punto, vamos ahora a explotar el hecho de que Seaborn se integre perfectamente con el uso de datasets de pandas.

Para ello, haremos uso del dataset *penguin*, que contiene datos sobre la taxonomía de pingüinos, para diferentes especies y islas recogidos en el archipiélago de Palmer. Una de las peculiaridades de este conjunto de datos, es que contiene algunos datos faltantes!

Pero esto no será un problema.

Para definir las gráficas en Seaborn, siempre tendremos que considerar algunas cosas a la hora de hacer uso de sus instrucciones:

- Variables a representar
- Conjunto de datos a utilizar

Ya en lugar de tener que filtrar el dataset previamente a la realización de los gráficos, como se ha venido haciendo hasta ahora en las dos anteriores lecciones, podemos definir a la hora de realizar las gráficas las variables llamándolas por el nombre  de la variable en el dataset.

Otra particularidad de Seaborn es que como está pensado para su uso de gráficas estadísticas, muchas veces se mostrarán por defecto las barras de error y los intervalos de confianza de los datos representados. Esto puede ser desactivado por elección de la usuaria.

### **Gráficas básicas** <a class="anchor" id="section_1_3_1"></a>

#### **Gráfica lineal**  <a class="anchor" id="section_1_3_1_1"></a>

Si queremos realizar una gráfica lineal conectando los diferentes puntos representados para las variables `x` e `y`, haciendo uso de seaborn podríamos utilizar el siguiente código como sintaxis general:

```python
    fig = sns.lineplot(x='variable_x', y='variable_y', data=dataset, ci=None, hue='opcional')
```

De esta forma en los argumentos del linepot, se deberia de especificar:

- `x`: Nombre de la variable (en el dataset) a dibujar en el eje X
- `y`: Nombre de la variable (en el dataset) a dibujar en el eje Y
- `data`: Nombre del dataset
- `ci`: Para mostrar el intervalo de confiaza de los datos, por defecto esta activado
- `hue`: Por si queremos utilizar una variable semántica para diferentes categorías

En este caso vamos a querer saber como cambia la longitud de la aleta en relación con la masa corporal. A priori, con lo aprendido hasta ahora podríamos pensar que la mejor opción es un *scatterplot* como aprendimos hasta ahora, pero hay un tipo de gráfica que no hemos aprendido porque en Matplotlib es un poco tediosa de hacer que es la gráfica de lineas. Veamos como hacerla: 

````python
# Iniciamos nuestro marco para la figura
plt.figure(figsize =(8,4))

# establecemos la gráfica
fig = sns.lineplot(x='body_mass_g' , y='bill_length_mm' , data=df_penguin, ci=None)

# cambiamos los nombres de los ejes. Es como el plt.ylabel y plt.xlabel que aprendimos hasta ahora. 
fig.set(xlabel='Masa corporal en g', ylabel='Longitud de la aleta en mm')

# ponemos ej titulo en la gráfica
fig.set_title('Gráfica lineal')
plt.show();
````

Aquí se puede ver la relación entre los pesos en gramos y la longuitud de las aletas de todos los datos de pingüinos disponibles, y vemos como parece que hay una tendencia que a mayor masa corporal mayor longitud de la aleta, lo cual de nuevo no nos debería de sorprender. 

Si ahora por ejemplo quisieramos mostrar las misma gráfica, pero diferenciando por especie o por lo sexo, con añadir una pequeña opción a lo anterior conseguiremos ese resultado!

Ya vereís como es mucho más rápido que utilizando Matplotlib directamente.Para ello añadimos el parámetro `hue`que se encargará de utilizar dicha variable semántica para determinar los colores de los diferentes elementos representados.

```python
    hue = 'variable_para_las_categorias'
```



````python
# creamos nuestra figura con el tamaño que queremos
plt.figure(figsize =(8,4))

# creamos el lineplot al que le añadimos un parámetro nuevo, el hue. En este caso, le estamos diciente que nos cree una línea para cada una de las especies que tenemos en el df
fig = sns.lineplot(x='body_mass_g' , y='bill_length_mm' , data = df_penguin, hue ='species')

# establecemos los nombres de los ejes
fig.set(xlabel='Masa corporal en g', ylabel = 'Longitud de la aleta en mm')

# establecemos un título para la gráfica
fig.set_title('Gráfica lineal')

# mostramos la gráfica
plt.show()

````

Como podéis ver, esto es mucho mas sencillo que en matplotlib y además nos ha generado la leyenda automaticamente!!! Perfecto! Fijaos que la leyenda nos la ha generado en la esquina superior izquierda, pero ... ¿podríamos moverla? La respuesta es sí y la veremos en el siguiente ejemplo usando el método `plt.legend` que ya conocemos. 

Por otro lado, fijaos como independiente de la especie, si que parece que se conserva el patrón de a mayor masa corporal mayor longitud de la aleta!

Hagamos otro ejemplo de esta gráfica, pero en este caso el hue lo haremos sobre la columa `sex`, ¿Qué conclusiones podéis sacar de la gráfica? 

````python
# establecemos el marco de nuestra gráfica
plt.figure(figsize =(8,4))

# hacemos el lineplot especificando el hue
fig = sns.lineplot(x='body_mass_g' , y='bill_length_mm' , data = df_penguin, ci=None, hue ='sex')

# cambiamos los nombres de los ejes
fig.set(xlabel='Masa corporal en g', ylabel = 'Longitud de la aleta en mm')

# ponemos título a la gráfica
fig.set_title('Gráfica lineal')

# cambiamos la posición de la leyenda usando el parámetro bbox_to_anchor, donde cada uno de los valores corresponden con la posición respecto a los ejes. Siendo 1 el límite de los ejes. 
plt.legend(bbox_to_anchor=(1.05, 1))

# mostramos la gráfica
plt.show()
````

#### Gráficas de dispersión <a class="anchor" id="section_1_3_1_2"></a>

Si ahora queremos realizar gráficas de dispersión o *scatterplots* podemos hacer uso de la instrucción

```python
    fig = sns.scatterplot(x='variable_x', y='variable_y', data=dataset, hue='opcional')
```

De esta forma en los argumentos del scatterplot, se debería de especificar:

- `x`: Nombre de la variable (en el dataset) a dibujar en el eje X
- `y`: Nombre de la variable (en el dataset) a dibujar en el eje Y
- `data`: Nombre del dataset
- `hue`: Por si queremos utilizar una variable semántica para diferentes categorías

````python
# iniciamos la figura
plt.figure(figsize =(8,4))

# usamos el método sns.scatterplot para indicar que tipo de figura queremos
fig = sns.scatterplot(x='body_mass_g' , y='bill_length_mm' , data = df_penguin)

# establecemos el título de la gráfica
fig.set_title('Gráfica de disperisón masa - longitud aleta')

# mostramos al gráfica
plt.show()
````

Fijaos que ahora hemos quitado el método `set` para especificar el nombre de los ejes y aún así nos aparecen a diferencia de Matplotlib. Y es que Seaborn tiene muchas ventajas,  y una de ellas es que nos pone los nombres de los ejes automáticamente, usando el nombre de las columnas que estamos visualizando. 

Si ahora añadieramos el argumento `hue`, podríamos diferenciar facilmente los datos representados por la especie, isla o sexo de los pingüinos.

````python
# Hacemos la gráfica con el parámetro hue para la columna de "sex"

# Iniciamos la gráfica
plt.figure(figsize =(8,4))

# Usamos el método scatterplot para visualizar los datos
fig = sns.scatterplot(x='body_mass_g' , y='bill_length_mm' , data = df_penguin, hue= 'species')

# cambiamos los nombres de los ejes
fig.set(xlabel='Masa corporal en g', ylabel = 'Longitud de la aleta en mm')

# ponemos título a la gráfica
fig.set_title('Gráfica de disperisón masa - longitud aleta por sexo')

# mostramos la gráfica
plt.show()
````

````python
# Hacemos la gráfica con el parámetro hue para la columna de "sex"

# Iniciamos la gráfica
plt.figure(figsize =(8,4))

# Usamos el método scatterplot para visualizar los datos
fig = sns.scatterplot(x='body_mass_g' , y='bill_length_mm' , data = df_penguin, hue= 'sex')

# cambiamos los nombres de los ejes
fig.set(xlabel='Masa corporal en g', ylabel = 'Longitud de la aleta en mm')

# ponemos título a la gráfica
fig.set_title('Gráfica de disperisón masa - longitud aleta por sexo')

# mostramos la gráfica
plt.show()
````

En este caso hemos podemos sugerir lo mismo que en el caso de la gráfica de lineas, y es que a mayor masa corporal mator longitud de la aleta. Vemos que independientemente de la gráfica, todos los caminos nos llevan a la misma conclusión.

#### Histogramas <a class="anchor" id="section_1_3_1_3"></a>

Si ahora queremos realizar histogramas con Seaborn podemos hacer uso de la instrucción

```python
    fig = sns.histplot(x='variable_x', data=dataset, hue='opcional', kde=True, bins=n_bins)
```

De esta forma en los argumentos del histograma, se debería de especificar:

- `x`: Nombre de la variable (en el dataset) a dibujar en el eje X
- `data`: Nombre del dataset
- `hue`: Por si queremos utilizar una variable semántica para diferentes categorías
- `kde`: Para mostrar la densidad de probabilidad
- `bins`: Número o lista de bins, es decir, de barritas que queremos en nuestra gráfica


En este caso querremos saber como se distribuye el peso de los pinguinos, es decir, ¿qué pesos de los que tenemos en nuestro conjunto de datos son los que más se repiten o los que más aparecen?

````python
# Histograma seleccionando un total de 15 bins
# iniciamos la figura
plt.figure(figsize =(8,4))

# elegimos el método hisplot con la variable que queremos visualizar
fig = sns.histplot(x='body_mass_g', data = df_penguin, bins = 15)

# establecemos el nombre de los ejes, pero recordad que esto no es obligatorio ya que Seaborn nos los genera automaticamente
fig.set(xlabel='Masa corporal en g', ylabel = 'Cuentas')

# Ponemos título a  la gráfica
fig.set_title('Histograma de los pesos de los pingüinos' )

# mostramos la gráfica
plt.show()

````

En este caso podemos ver que hay muchos más pingüinos entre los 3 y 4 kg que pingüinos con más de 5 kg de peso. Imaginaos ahora que nos planteamos, si este patrón que hemos observado de forma general es algo común para todas las islas o sexos, es decir: 

- ¿Pesan más los pingüinos de una isla que de otra?

- ¿Pesan más los machos que las hembras?

````python
#  Histograma seleccionando como variable semantica la variable isla y un total de 20 bins, para saber si pesan más en una isla que otra

# iniciamos la gráfica
plt.figure(figsize =(8,4))

# establecemos la gráfica con elm método hue para la columna de "island"
fig = sns.histplot(x='body_mass_g', data = df_penguin, hue = 'island', bins = 20)

# cambiamos el nombre de los ejes
fig.set(xlabel='Masa corporal en g', ylabel = 'Cuentas')

# ponemos título a la gráfica
fig.set_title('Histograma de los pesos de los pingüinos, en función de la isla')

# mostramos la gráfica
plt.show()
````

Viendo la gráfica, podemos ver un resultado claro y es que los resultados sugieren que los pingüinos de la isla de Biscoe pesan mñas que los de las otras islas. Por el contrario, si nos fijamos en los de las otras islas esto no parece tan claro, pareciendo que en ambas islas todos pesan similar. Lo que si podemos concluir es que en la isla de Dream hay más pingüinos que en la de Torgensen, ya que las barras para Dream son más altas. Recordemos que el histograma nos va a mostrar la frecuencia en la que aparecen animales en cada isla. 

Veamos que pasa ahora con el sexo:

````python
#  Histograma seleccionando como variable semantica la variable sex, un total de 20 bins y con KDE, para responder a la pregunta de si hay diferencias entre sexos en el peso corporal

# iniciamos la figura
plt.figure(figsize =(8,4))

# seleccionamos el histplot y le añadimos el parámetro kde que nos va a mostrar una línea de tendencia 
fig = sns.histplot(x='body_mass_g', data = df_penguin, hue = 'sex', kde=True)

# cambiamos el nombre de los ejes
fig.set(xlabel='Masa corporal en g', ylabel = 'Cuentas')

# ponemos título a la gráfica
fig.set_title('Histograma de los pesos de los pingüinos, en función del sexo, con su densidad de probabilidad')

# mostramos la gráfica
plt.show()
````

En este caso podemos ver como las hembras tienden a pesar menos que los machos, resultado que no es sorpresa, ya que es algo que nos podríamos esperar. ¿Qué pasaría si vemos ahora el número de pingüinos de cada especie que hay en cada isla? 

````python
#  Histograma seleccionando como variable  la variable isla y un total de 3 bins

# iniciamos la gráfica
plt.figure(figsize =(8,4))

# establecemos el gráfico
fig = sns.histplot(x='species', data = df_penguin,hue = 'island')

# cambiamos el nombre de los ejes
fig.set(xlabel='Species', ylabel = 'Cuentas')

# ponemos el título a la gráfica
fig.set_title('Histograma de la cantidad de pingüinos, en función de la isla')

# mostramos la gráfica
plt.show()
````

Con esta gráfica podemos concluir que en la especide de Gentoo solo aparece en la isla de Biscoe,, la especie Chinstrap aparece solo en Dream y que la especie de Adelie aparece en las en las tres islas, siendo más abundante en la Dream, luego en Torgensen y por último en la de Biscoe

#### Diagramas de caja o *boxplot*<a class="anchor" id="section_1_3_1_4"></a>

Si ahora queremos realizar diagramas de caja utilizando Seaborn, podemos hacer uso de la instrucción:

```python
    fig = sns.boxplot(x='variable_x', data=dataset, hue='opcional', kde=True, bins=n_bins)
```

De esta forma en los argumentos del *boxplot*, se debería de especificar:

- `x`: Nombre de la  columna (en el dataset) a dibujar en el eje X
- `data`: Nombre del dataset
- `hue`: añadimos variable categórica 


En el caso de los boxplots en Seaborn, por defecto vienen con orientacion horizontal. De forma que para que el boxplot sea representado de forma vertical, deberemos escribir el nombre de la variable a representar en el eje y de los argumentos de boxplot.

El hecho de que esto funcione así, nos permitirá facilmente representar boxplots haciendo uso de variables categóricas para ver con más detalle como están distribuidos los datos.

Lo ilustraremos con los siguientes ejemplos:

````python
# Boxplot por defecto 
# inciamos la figura
plt.figure(figsize =(8,4))

# elegimos el método boxplot y especificamos la columna que queremos usar
fig = sns.boxplot( x ='bill_length_mm', data = df_penguin)

# cambiamos el nombre del eje x. Recordad que este no es obligatorio en seaborn ya que nos lo genera automaticamente. 
fig.set(xlabel='Longitud de la aleta en mm')

# ponemos título a la gráfica
fig.set_title('Boxplot de la longitud de la aleta')

# mostramos la gráfica
plt.show()
````

````python
# Boxplot por vertical 

#iniciamos la figura
plt.figure(figsize =(8,4))

# creamos nuestro boxplot, fijaos que ahora hemos puesto el parámetro y en vez de x para que nos aparezca en el otro sentido
fig = sns.boxplot( y ='bill_length_mm', data = df_penguin)

# ponemos nombre al eje
fig.set(ylabel='Longitud de la aleta en mm')

# ponemos nombre a la figura
fig.set_title('Boxplot de la longitud de la aleta')

# mostramos la figura
plt.show()
````

Imaginemos ahora que queremos ver esta misma gráfica, pero que queremos tener un *boxplot* para cada una de las islas, es decir, en este caso tendremos en el eje de las `x` las islas y en el de la `y` la longitud del culmen (`bill_length_mm`), en este caso tendremos que usar el parámetro `x` e `y` en nuestro gráfico. Vemamos un ejemplo:

````python
df_penguin.head()
````

````python
# Boxplot por vertical, por isla
# iniciamos la figura con un tamaño especificado
plt.figure(figsize =(8,4))

# # establecemos los parámetros del boxplot, especificando que en el eje de las x queremos las islas y en el de las y la aleta
fig = sns.boxplot(x= 'island', y ='flipper_length_mm', data = df_penguin)

# cambiamos el nombre de los ejes
fig.set(xlabel='Isla', ylabel= 'Longitud de la aleta en mm')

# ponemos título a la gráfica
fig.set_title('Boxplot de la longitud de la aleta en mm por isla')

# mostramos la gráfica
plt.show()

````

En este caso podemos ver como las aletas de los pingüinos de la isla de Biscoe tienden a ser más grandes que las aletas en las otras islas. Por otro lado, no podríamos decir claramente que entre las islas de Torgensen y Dream exista una clara diferencia entre la longitud de sus aletas, ya que si comparamos  sus medianas (las líneas negras de dentro de las cajas).


Ahora imaginemos que queremos añadir a nuestra figura sobre las especies, en este caso añadiremos el parámetro `hue`. 

````python
# Boxplot por vertical, por isla

# iniciamos la gráfica
plt.figure(figsize =(8,4))

# iniciamos el boxplot y añadimos el parámetro hue
fig = sns.boxplot(x= 'island', y ='body_mass_g',hue= 'species', data = df_penguin)

# ponemos nombre a los ejes
fig.set(xlabel='Isla', ylabel= 'Longitud de la aleta en mm por isla y sexo' )

# ponemos título
fig.set_title('Boxplot de la longitud de la aleta')

# mostramos la gráfica
plt.show()
````

En este caso, podemos ver que en la isla *Torgersen*, únicamente tenemos pingüinos de la especie Adelie o que en la isla *Dream*, tenemos algunos datos potencialmente fuera de lo común o erróneos. #### Diagrama de enjambre <a class="anchor" id="section_1_3_2_1"></a>


Hemos visto que los *scatterplots* son uno de los gráficos más básicos y útiles que se pueden utilizar, y los hemos estado usando mucho durante estas sesiones. En seaborn podemos llevar estos gráficos un paso más allá y usarlos para representar valores categóricos. Para eso usaremos un *swarmplot* es un tipo de gráfico de dispersión que se utiliza para representar valores categóricos. Es muy similar al diagrama de barras, pero evita la superposición de puntos, lo cual nos ayuda también a ver cómo están distribuidos los puntos en el rango de valores de dichos datos.

Para hacer este tipo de gráficas, utilizaremos la siguiente sintaxis:

```python
    fig = sns.swarmplot(x='variable_x', y= 'variable_y', data = dataset, hue='opcional')
```

De esta forma en los argumentos del swarmplot, se debería de especificar:

- `x`: Nombre de la variable (o columna) a dibujar en el eje X
- `y`: Nombre de la variable (o columna) a dibujar en el eje Y
- `data`: Nombre del dataset
- `hue`: para añadir valores categóricos en nuestra visualización

En este caso, nos vamos a preguntar en que isla pesan más los pingüinos y como se distribuyen los pesos dentro de cada isla, es decir, dentro de cada isla, que pesos son los que más se repiten y los que menos. 

````python
# Diagrama de enjambre del peso de los pingüinos por isla

# iniciamos la figura
plt.figure(figsize =(8,4))

# usamos el swarmplot para visualizar nuestra pregunta
fig = sns.swarmplot(x='island', y = 'body_mass_g' ,data = df_penguin)

# ponemos nombre a los ejes
fig.set(xlabel='Isla', ylabel= 'Peso de los pingüinos por isla' )

# ponemos título a la gráfica
fig.set_title('Diagrama de enjambre del peso de los pingüinos por isla')

# mostramos la gráfica
plt.show()
````

Para empezar, tenemos que ver en esta gráfica, cada punto corresponde con un pingüino. De la gráfica anterior se puede extraer a simple vista que los pesos de los pingüinos de la isla Biscoe, son por lo general más altos. Además podemos ver que, por ejemplo en la isla de Biscoe la parte más ancha está en torno a los 5000 gr de peso, es decir, la mayoría de nuestos pingüinos pesan en torno a esto, sería como nuestra mediana!

Ahora vamos un paso más allá y diferenciamos por el sexo de los pingüinos:

````python
# Diagrama de enjambre del peso de los pingüinos por isla y sexo

# iniciamos la figura
plt.figure(figsize =(8,4))

# usamos el parámetro hue para añadir la variable categórica de sexo
fig = sns.swarmplot(x='island', y = 'body_mass_g' ,hue='sex', data = df_penguin)

# ponemos nombre al eje
fig.set(xlabel='Isla', ylabel= 'Peso de los pingüinos por isla' )

# título de gráfica
fig.set_title('Diagrama de enjambre del  peso de los pingüinos por isla y sexo')

# mostramos la gráfica
plt.show()
````

Haciendo uso de las variables categóricas disponibles hemos podido entender mejor cómo están distribuidos los datos de los pesos de los pingüinos, para las diferentes islas, así como los diferentes sexos. De la gráfica anterior, podemos extraer que los pingüinos macho suelen tener un peso mayor que en el caso de las hembras y que el peso medio (donde mayor cantidad de puntos hay en las gráficas), es superior en todos los casos para ambos sexos en el caso de la isla de Biscoe.

#### Countplot <a class="anchor" id="section_1_3_2_2"></a>

Como se ha visto hasta ahora, cuando buscamos realizar histogramas o diagramas de barras, siempre hemos tenido que trabajar con variables de tipo numérico.
No obstante en Seaborn existe una función que nos permitirá hacer esto mismo utilizando las variables categóricas del dataset con el que estemos trabajando.

Para ello podremos hacer uso de la instrucción countplot, que tiene la siguiente sintaxis:

```python
    fig = sns.countplot(x='variable_x', data = dataset)
```

De esta forma en los argumentos del countplot, se deberia de especificar:

- `x`: Nombre de la variable (o columna del dataset) a dibujar en el eje X
- `y`: Nombre de la variable (o columna del dataset) a dibujar en el eje Y
- `data`: Nombre del dataset
- `hue`: Por si queremos utilizar una variable semántica para diferentes categorías

Tendremos que escoger únicamente `x` o `y`, para especificar si queremos representar los datos en el eje de las X o Y.

En este contestaremos a la siguiente pregunta, en ¿qué isla tenemos más pingüinos?

````python
# iniciamos la figura con un tamaño especifico
plt.figure(figsize =(8,4))

# iniciamos nuestra gráfica para contestar a la pregunta
fig = sns.countplot(x='island', data = df_penguin)

# cambiamos el nombre de los ejes
fig.set(xlabel='Isla', ylabel= 'Cuentas' )

# cambiamos el título de la gráfica
fig.set_title('Número de pingüinos por isla')

# motramos la gráfica
plt.show()
````

De esta forma hemos obtenido rápidamente el valor total de de pingüinos por cada una de las islas, como si se tratase de una variable numérica.
Si ahora adicionalmente añadiéramos algún valor al atributo `hue` en la gráfica anterior:

````python
# iniciamos la figura con un tamaño especifico
plt.figure(figsize =(8,4))

# iniciamos nuestra gráfica para contestar a la pregunta, añadiendo sex al parámetro hue
fig = sns.countplot(x='island', hue = "sex", data = df_penguin)

# cambiamos el nombre de los ejes
fig.set(xlabel='Isla', ylabel= 'Cuentas' )

# cambiamos el título de la gráfica
fig.set_title('Número de pingüinos por isla')

# motramos la gráfica
plt.show()
````

Ahora ya conocemos con todo lujo de detakke, la cantidad de pingüinos que hay por cada sexo en cada una de las islas. Como se puede ver, tener la capacidad de representar variables categóricas es extremadamente útiles para conocer en detalle los datos que se disponen.

#### Catplot <a class="anchor" id="section_1_3_2_3"></a>
 
Si ahora queremos saber cómo se pueden relacionar una variable categórica con una variable de tipo numérico podemos hacer uso de la instrucción `Catplot`.

La sintaxis general para este tipo de gráficas es la siguiente:

```python
    fig = sns.catplot(x='variable_x', y ='variable_y', data=dataset)
```

De esta forma en los argumentos del countplot, se deberia de especificar:

- `x`: Nombre de la variable (en el dataset) a dibujar en el eje X

- `y`: Nombre de la variable (en el dataset) a dibujar en el eje Y

- `data`: Nombre del dataset

- `hue`: Por si queremos utilizar una variable semántica para diferentes categorías

**NOTA**: No podemos añadir el título exactamente igual que en el resto de los casos para ello se debe hacer uso de `plt.title()`. Además, es similar a un `swarmplot` pero sin visualizar la mediana de los datos. 

De esta forma vamos a ilustrarlo con un ejemplo:

````python
# iniciamos la figura con el tamaño que especificamos
plt.figure(figsize =(10,15))

# iniciamos nuestra gráfica
fig = sns.catplot(x='island', y ='flipper_length_mm', data = df_penguin)

# # cambiamos el nombre de los ejes
fig.set(xlabel='Isla', ylabel= 'Longitud de la aleta en mm' )

# ponemos título a la gráfica
plt.title('Pingüinos')

# mostramos la gráfica
plt.show();
````

De la gráfica anterior podemos extraer las diferentes longitudes de las aletas por cada una de las islas y ver donde estarían representados cada uno de los valores en la gráfica. De nuevo, vemos que los pingüinos de la isla de Biscoe tiene unos tamaños de aletas más grandes que en el resto de las islas, mientras que en las dos otras islas no se puede observar una diferencia clara. 

Si ahora adicionalemente quisieramos saber cómo está separado por especies, añadiendo el `hue` podríamos conseguirlo rápidamente.

````python
# iniciamos la figura con el tamaño que especificamos
plt.figure(figsize =(10,15))

# iniciamos nuestra gráfica
fig = sns.catplot(x='island', y ='flipper_length_mm', data = df_penguin, hue = "species")

# # cambiamos el nombre de los ejes
fig.set(xlabel='Isla', ylabel= 'Longitud de la aleta en mm' )

# ponemos título a la gráfica
plt.title('Pingüinos')

# mostramos la gráfica
plt.show();
````

Ahora ya no solo vemos como se distribuyen los datos entre islas sino también por especies! Y como hemos estado viendo, en la isla de Torgensen solo tenemos la especie Adelie, en la de Biscoe las especies de Gentoo y Adelie y en Dream Adelie y Chinstrap. 

Además, podemos concluir que, en la isla de Biscoe son los pingüinos de la especie Gentoo es la que presenta las longitudes de aleta más grande. Por otro lado, vemos que en la isla de Dream, los puntos aparecen bastante superpuestos, lo que nos podría indicar que no hay diferencia entre la longitud de las aletas entre especies en esa isla.

#### Pairplot <a class="anchor" id="section_1_3_2_4"></a>

Este tipo de gráficas nos permitirá obtener los histogramas y diagramas de dispersión de todas las variables numéricas de las que disponga el dataset con el que estemos trabajando. Nos van a servir para empezar a comprender de forma visual como se distribuyen los datos.

De esta forma, para realizar esto emplearemos la siguiente instrucción:

```python
    fig = sns.pairplot(data=dataset, hue='opcional', kind='opcional')
```

De esta forma en los argumentos del countplot, se debería de especificar:

- `data`: Nombre del dataset
- `hue`: Por si queremos utilizar una variable semántica para diferentes categorías
- `kind`: {scatter, kde, hist, reg}, por defecto viene scatter, que mostrará en las gráficas no diagonales los diagramas de dispersión


````python
# iniciamos la figura con una tamaño de figura especificado
plt.figure(figsize =(8,4))

# iniciamos la figura del pair plot y le pasamos todo el df
sns.pairplot(data = df_penguin)

# mostramos la figura
plt.show()
````

Entender estas gráficas puede ser algo más complejo y la explicación necesita mucho detalle, es por esto, que esta gráfica será interpretada durante la clase de mañana. De todas formas, si nos fijamos aparecen dos tipos de gráficas que ya hemos aprendido y que mirando con detenimiento podríamos llegar a entender, en concreto tenemos: 

- `*scatterplots*`  que nos muestran la relación entre dos variables numéricas.

- `histograma` que nos muestra la distribución de nuestras variables. 

También podemos utilizar las variables categóricas disponibles, para ver cómo estarían distribuidos los datos.
Para ello de nuevo hacemos uso de la instrucción hue.

````python
# iniciamos la figura con una tamaño de figura especificado
plt.figure(figsize =(8,4))

# iniciamos la figura del pair plot y le pasamos todo el df
sns.pairplot(data = df_penguin, hue = "species")

# mostramos la figura
plt.show()
````

Ahora en este caso, en la diagonal en lugar de tener el histograma nos mostrará la densidad de probabilidad. Ya se explicará en futuras lecciones exactamente qué significa esto. Ahora mismo con saber que representa las probabilidades donde podemos encontrar ciertos valores en la gráfica, será suficiente.
